```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
import statsmodels.api as sm
sns.set_style("darkgrid")
mpl.rcParams['figure.figsize'] = (10,5)

original = pd.read_csv('original.csv',index_col=0)
remake = pd.read_csv('remake.csv',index_col=0)
```

```python
remake
```

```python
print(original.imdb.mean())
print(original.metascore.mean())
print(original.rotten_tomatoes.mean())
print(original.profit.mean())
```

```python
print(remake.imdb.mean())
print(remake.metascore.mean())
print(remake.rotten_tomatoes.mean())
print(remake.profit.mean())
```

```python
original.plot(kind='box')
original.plot(kind='line')
remake.plot(kind='box')
remake.plot(kind='line')
plt.show()
```

```python
ori_list = list(original.columns)
ori_list.remove('title')

for item in ori_list:
    ax1 = original[item].plot()
    print(type(ax1))
    ax2 = ax1.twinx()
    ax2.plot(original.index,color='teal')
    plt.title(item)
    plt.show()
```

```python
re_list = list(remake.columns)
re_list.remove("title")
print(re_list)

for item in re_list:
    ax1 = remake[item].plot()
    print(type(ax1))
    ax2 = ax1.twinx()
    ax2.plot(remake.index, color='teal')
    plt.title(item)
    plt.show()
```

```python
sns.heatmap(original.corr(),annot=True)
```

```python
sns.heatmap(remake.corr(),annot=True)
```

```python
indep_ori = original.iloc[:,[1,2,4,5,6,7]].astype(float)
dep_ori = original.iloc[:,3].astype(float)

indep_ori = sm.add_constant(indep_ori).fillna(0)

regression_model_ori = sm.OLS(dep_ori, indep_ori).fit()
print(regression_model_ori.summary())
```

```python
indep_re = remake.iloc[:,[1,2,4,5,6,7]].astype(float)
dep_re = remake.iloc[:,3].astype(float)

indep_re = sm.add_constant(indep_re).fillna(0)

regression_model_re = sm.OLS(dep_re, indep_re).fit()
print(regression_model_re.summary())
```
