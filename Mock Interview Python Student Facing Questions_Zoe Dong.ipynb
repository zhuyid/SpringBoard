<b> Q5. Which Job Role has the highest <i> proportion </i> of individuals who earn >50K? </b>

<b> Put your code below </b>

```python
roles = df[['Job-Role','Earnings']]
roles = roles.reindex(columns = roles.columns.tolist() + ['Earnings_0 or 1'])
roles
```

```python
for i, row in roles.iterrows():
    if roles.iloc[i, 1] == ' <=50K':
        roles.iloc[i,2] = 0
    else:
        roles.iloc[i,2] = 1
```

```python
roles = roles.iloc[:,[0,2]]
roles
```

```python
earnings = roles.groupby(['Job-Role']).mean()
earnings
```

```python
earnings.plot(kind='bar',title = 'The Proportion of Individuals Who Earn >50K')
```

# Answer: Exec-managerial has the highest proportion of individuals who earn >50K.
