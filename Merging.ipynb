```python
import pandas as pd

df1 = pd.read_csv("released.csv")
df2 = pd.read_csv("upcoming.csv")
df3 = pd.read_csv("disney.csv",index_col=0)
```

```python
df = pd.merge(df1, df3, left_on='title',right_on='title')
original =  df[df['release_date_y']<'2010-01-01'].iloc[3:]
original = original.drop(columns = ["Unnamed: 0","original_film_title","running_time","release_date_x"])
original.set_index('release_date_y',inplace=True)
original['rotten_tomatoes'] = original['rotten_tomatoes'].str.replace('%','')
original.index.names=['release_date']
original['profit'] = original['box_office'] - original['budget']
original['pct_change'] = original['profit'].pct_change()
```

```python
remake = df[df['release_date_y']>'2010-01-01']
remake = remake.drop(columns = ['Unnamed: 0','original_film_title','running_time','release_date_y'])
remake.set_index('release_date_x',inplace=True)
remake['rotten_tomatoes'] = remake['rotten_tomatoes'].str.replace('%','')
remake.index.names=['release_date']
remake['profit'] = remake['box_office'] - remake['budget']
remake['pct_change'] = remake['profit'].pct_change()
```

```python
original.to_csv('original.csv')
remake.to_csv('remake.csv')
```
